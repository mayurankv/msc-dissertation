```python
from stochastic_volatility_models.src.data.realised_measures import get_realised_measure_maximum
from pandas import Index

# v_0, theta,kappa,eta,rho
# 0.01417449066157833,0.052396679044242694,5.198158055919218,0.45478037502723,-0.9123537255278572
time = "2017-11-30"
nb = Notebook(
	time=time,
	model=Heston(
		parameters={"initial_variance": 0.01417449066157833, "long_term_variance": 0.2066183732051921, "volatility_of_volatility": 0.014573181037181637, "mean_reversion_rate": 2.558439100426357, "wiener_correlation": 0.9314400109868923},
	),
)
nb.model.fix_initial_variance(nb.underlying.realised_volatility(Index([nb.time.astype(str)])).iloc[0])
nb.model.set_upper_long_term_variance(get_realised_measure_maximum(nb.underlying.ticker))
```
